```python
# 상승장 하락장

# 이동 평균선 사용 : 이동 평균 값보다 현재가가 높다면 상승장, 낮으면 하락장

btc = pbit.get_ohlcv()
close = btc['close']
#print(close)

# 수동으로 이동평균 계산
avg1 = sum(close[0:5])/5
avg2 = sum(close[5:10])/5
avg3 = sum(close[10:15])/5
print(avg1, avg2, avg3)
```

```python
# 자동으로 이동평균 계산

btc = pbit.get_ohlcv()
close = btc['close']

roll5 = close.rolling(5)
mean5 = roll5.mean()
print(mean5)

# 5일간의 평균을 계산하기 때문에 최초 4일은 계산 값이 없이 NaN으로 표시
```

```python
# 상승장 하락장 파악
btc = pbit.get_ohlcv()
close = btc['close']

roll5 = close.rolling(5)
mean5 = roll5.mean()
last_mean5 = mean5[-2]

# 비트 코인 현재 값
price = pbit.get_current_price('BTC')

print('상승장') if price > last_mean5 else print('하락장')
```
